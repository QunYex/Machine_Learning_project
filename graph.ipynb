```python
from sklearn.metrics.pairwise import pairwise_distances

vars_to_use = ['b','e','AC','FM','UC','DL','DS','DP','DR','LB','AC.1','FM.1','UC.1','DL.1','DS.1','DP.1','ASTV','MSTV','ALTV','MLTV','Width','Min','Max','Nmax','Nzeros','Mode','Mean','Median','Variance','Tendency']# pick vars

xdata = pairwise_distances(data[vars_to_use].values, metric='correlation')
sns.heatmap(xdata, cmap=cmap, annot=False)
print('What is wrong with this plot?')
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

xdata = pairwise_distances(std.fit_transform(data[vars_to_use].values), metric='correlation')
sns.heatmap(xdata, cmap=cmap, annot=False)
print('Is there still something worng?')
```

```python
f, ax = plt.subplots(figsize=(8, 7))


std = StandardScaler()

data_copy = data.copy().sort_values(by=['AC','LB','NSP'])

xdata = pairwise_distances(std.fit_transform(data_copy[vars_to_use].values), metric='correlation')
sns.heatmap(xdata, cmap=cmap, annot=False)
print('Is there anything we can conclude?')
```
